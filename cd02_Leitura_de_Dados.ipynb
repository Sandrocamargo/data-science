# AULA 2 - COLETA E LEITURA DE DADOS


```python
# Bibliotecas necessárias
import pandas as pd
import requests

print("Ambiente pronto!\n")
```

# 1. LEITURA DE CSV

```python
print("=== LEITURA DE CSV ===")

# Exemplo com dataset público (Iris)
url_csv = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

df_csv = pd.read_csv(url_csv)

print("Primeiras linhas:")
print(df_csv.head())

print("\nInformações do dataset:")
print(df_csv.info())

print("\nEstatísticas básicas:")
print(df_csv.describe())
```

# 2. LEITURA DE EXCEL

```python
print("\n=== LEITURA DE EXCEL ===")

# Criando um exemplo simples (caso não tenha arquivo local)
df_exemplo = pd.DataFrame({
    'Nome': ['Ana', 'Bruno', 'Carlos'],
    'Idade': [23, 35, 29],
    'Nota': [8.5, 7.0, 9.2]
})

# Salvando e lendo Excel
df_exemplo.to_excel("exemplo.xlsx", index=False)

df_excel = pd.read_excel("exemplo.xlsx")

print(df_excel)
```

# 3. LEITURA DE JSON

```python
print("\n=== LEITURA DE JSON ===")

# Exemplo de JSON
dados_json = [
    {"nome": "Ana", "idade": 23},
    {"nome": "Bruno", "idade": 35}
]

df_json = pd.DataFrame(dados_json)

print(df_json)

url = 'https://microsoftedge.github.io/Demos/json-dummy-data/64KB.json'
df_json2 = pd.read_json(url)
print(df_json2)
```

# 4. CONSUMO DE API

```python
print("\n=== CONSUMO DE API ===")

# API pública (JSONPlaceholder - fake API)
url_api = "https://jsonplaceholder.typicode.com/posts"

response = requests.get(url_api)

if response.status_code == 200:
    dados_api = response.json()
    df_api = pd.DataFrame(dados_api)

    print("Primeiras linhas da API:")
    print(df_api.head())
else:
    print("Erro ao acessar API")
```

# 5. INSPEÇÃO INICIAL DOS DADOS

```python
print("\n=== INSPEÇÃO DOS DADOS ===")

print("Colunas:")
print(df_csv.columns)

print("\nTipos de dados:")
print(df_csv.dtypes)

print("\nValores ausentes:")
print(df_csv.isnull().sum())
```
