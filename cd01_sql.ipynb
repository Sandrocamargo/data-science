```python
import sqlite3
import pandas as pd
import requests
import os

print("\n=== INICIANDO AULA ===\n")
```

# 1. BAIXAR BANCO (SE NÃO EXISTIR)

```python
db_file = "chinook.db"

if not os.path.exists(db_file):
    print("Baixando banco de dados Chinook...")

    url = "https://github.com/lerocha/chinook-database/raw/master/ChinookDatabase/DataSources/Chinook_Sqlite.sqlite"
    response = requests.get(url)

    with open(db_file, "wb") as f:
        f.write(response.content)

    print("Download concluído!\n")
else:
    print("Banco já existe.\n")
```

# 2. CONECTAR AO BANCO

```python
conn = sqlite3.connect(db_file)
print("Conectado ao banco!\n")
```

# 3. LISTAR TABELAS

```python
print("=== TABELAS DISPONÍVEIS ===")

query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql(query, conn)

print(tables, "\n")
```

# 4. SELECT BÁSICO


```python
print("=== SELECT (PRIMEIRAS LINHAS) ===")

query = "SELECT * FROM customer LIMIT 5;"
df = pd.read_sql(query, conn)

print(df, "\n")
```

```python
print("=== SELECT (COLUNAS ESPECÍFICAS DAS PRIMEIRAS LINHAS) ===")

query = "SELECT FirstName, LastName, City, State, Country FROM customer LIMIT 10;"
df = pd.read_sql(query, conn)

print(df, "\n")
```

# 5. FILTRO (WHERE)

```python
print("=== FILTRO: CLIENTES DO BRASIL ===")

query = "SELECT * FROM customer WHERE Country = 'Brazil';"
df = pd.read_sql(query, conn)

print(df, "\n")
```

# 6. AGRUPAMENTO (GROUP BY)

```python
print("=== TOTAL DE CLIENTES POR PAÍS ===")

query = """
SELECT Country, COUNT(*) as total_clientes
FROM customer
GROUP BY Country
ORDER BY total_clientes DESC;
"""

df = pd.read_sql(query, conn)
print(df, "\n")
```

# 7. JOIN

```python
print("=== JOIN: CLIENTES E FATURAS ===")

query = """
SELECT c.FirstName, c.LastName, i.InvoiceId, i.Total
FROM customer c
JOIN invoice i ON c.CustomerId = i.CustomerId
LIMIT 10;
"""

df = pd.read_sql(query, conn)
print(df, "\n")
```

# 8. AGREGAÇÃO (SUM)

```python
print("=== RECEITA POR PAÍS ===")

query = """
SELECT c.Country, SUM(i.Total) as receita_total
FROM customer c
JOIN invoice i ON c.CustomerId = i.CustomerId
GROUP BY c.Country
ORDER BY receita_total DESC;
"""

df = pd.read_sql(query, conn)
print(df, "\n")
```

# 9. CONSULTA MAIS COMPLEXA

```python
print("=== CLIENTE QUE MAIS GASTOU ===")

query = """
SELECT c.FirstName, c.LastName, SUM(i.Total) as total_gasto
FROM customer c
JOIN invoice i ON c.CustomerId = i.CustomerId
GROUP BY c.CustomerId
ORDER BY total_gasto DESC
LIMIT 1;
"""

df = pd.read_sql(query, conn)
print(df, "\n")
```
